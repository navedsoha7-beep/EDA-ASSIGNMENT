```python
#1. Read the Bike Details dataset into a Pandas DataFrame and display its first 10 rows.

import pandas as pd
df=pd.read_csv('BIKE DETAILS.csv/BIKE DETAILS.csv')


print(df.head(10))
```

```python
#2.  Check for missing values in all columns and describe your approach for handling them.


missing_values=df.isnull().sum()
print(missing_values)


#null in most of the cases
```

```python
#3. Plot the distribution of selling prices using a histogram and describe the overall trend.

import matplotlib.pyplot as plt
plt.hist(df['selling_price'],bins=5,color='black',edgecolor='red')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.title('Distribution of Selling Prices Of Bikes')
plt.show()
```

```python
#4.  Create a bar plot to visualize the average selling price for each seller_type and write one observation.


average_selling_price=df.groupby('seller_type')['selling_price'].mean()
plt.bar(average_selling_price.index,average_selling_price.values,color='skyblue',edgecolor='magenta')
plt.xlabel('Seller Type')
plt.ylabel('Average Selling Price')
plt.title('Average Selling Price For Each Seller Type')
plt.show()
#
```

```python
#5.  Compute the average km_driven for each ownership type (1st owner, 2nd owner, etc.), and present the result as a bar plot.


average_km_driven_by_owners_of_car=df.groupby('owner')['km_driven'].mean()
plt.bar(average_km_driven_by_owners_of_car.index,average_km_driven_by_owners_of_car.values,color='purple',edgecolor='yellow')
plt.xlabel(' Both Owners')
plt.ylabel('Average Km Driven')
plt.title('Average Km Driven For Each Owner')
plt.show()
```

```python
#6. Use the IQR method to detect and remove outliers from the km_driven column. Show before-and-after summary statistics.

Q1_km_drived=df['km_driven'].quantile(0.25)
Q3_km_drived=df['km_driven'].quantile(0.75)
IQR_km_drived=Q3_km_drived-Q1_km_drived

outliers_km_drived=df[df['km_driven']>Q3_km_drived+1.5*IQR_km_drived]
print(outliers_km_drived)
```

```python
#7. Create a scatter plot of year vs. selling_price to explore the relationship between a bike's age and its price.
plt.scatter('x ', 'y')
plt.xlabel('YEAR')
plt.ylabel('SELLING PRICE OF BIKES')
plt.title('YEAR VS SELLING PRICE OF BIKES')
plt.show()
```

```python
#8. Convert the seller_type column into numeric format using one-hot encoding. Display the first 5 rows of the resulting DataFrame.

df_encoded=pd.get_dummies(df,columns=['seller_type'])
print(df_encoded.head(5))
```

```python
#9.  Generate a heatmap of the correlation matrix for all numeric columns. What correlations stand out the most?


import seaborn as sns

numeric_columns=df.select_dtypes(include=['number'])

sns.heatmap(numeric_columns.corr(),annot=True,cmap='cool')
plt.show()

```
